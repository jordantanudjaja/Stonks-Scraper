<h1>Installing and importing fbprophet and other necessary libraries</h1>

<p>Prophet is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonal effects and several seasons of historical data. Prophet is robust to missing data and shifts in the trend, and typically handles outliers well.</p>

<p>Prophet is open source software released by Facebook's Core Data Science team .

Full documentation and examples available at the homepage: https://facebook.github.io/prophet/</p>

<h1>Pre-processing the data to fit the fbprophet model format</h1>

```python
df.columns = ['ds', 'y']
df.head()
```

<h1>Initializing the Prophet model and fitting the Time-series data</h1>

```python
model = Prophet(daily_seasonality = False, weekly_seasonality = False) # For data involving securities that span for years, I believe
                                                                       # daily and weekly seasonality will skew the data and so I deactivated
                                                                       # both of these parameters
```

```python
model.fit(df)
```

```python
model.component_modes
```

<h1>Forecasting the closing prices of the security up to a year's worth of data</h1>

```python
future_dates = model.make_future_dataframe(periods = 365) # 365 days to forecast for the next year
prediction = model.predict(future_dates)
model.plot(prediction)
```

```python
# Showing the general trend line across the year and through each month
model.plot_components(prediction)
```

<h1>Doing a Cross-validation analysis to see how well the model fits the data</h1>

```python
help(cross_validation)
```

```python
# Doing cross-validation analysis to see how accurate my model is 
# Explanation of parameters in the cross_validation method:
# Horizon: The size of the testing dataset. This period begins from (latest date - Horizon) to the latest date
# Initial: The size of the training dataset. This period begins from First date in the dataframe and ends in (First date + Initial)
# The validation dataset is thus the period from (First date + Initial) to (Latest date - Horizon)\
# Based on this logic, the larger your training dataset (Initial value), the more data your model will fit in, and
# probably the smaller the MSE

df_cv = cross_validation(model, horizon = "365 days", initial = "1095 days") # Change the Initial Value to how you see fits best 
                                                                             # with the security's historical data
df_cv.head()
```

```python
# Displaying the performance metrics table of the testing dataset
df_perf = performance_metrics(df_cv)
df_perf.head()
```

```python
# Plotting the range of Root Mean Squared Error (RMSE) and comparing it with the security's closing price
# values for the last year

# Remember that RMSE has the same unit as the Dependent Variable, in this case the closing price
# of the security, and thus a good fit for our model will show the values of our RMSE trend line
# to be way below the values of the range of closing prices of the test dataset (Horizon parameter)

fig1 = plot_cross_validation_metric(df_cv, metric = "rmse", figsize = (12, 12))

working_days_last_year = 253 # This number should result in the first row of the dataframe below
                             # to match the first date of the testing dataset
fig2 = df.tail(working_days_last_year).plot(x = 'ds', y = 'y', kind = "line", xlabel = "Date", ylabel = "Closing Prices", figsize = (12, 12))
```
